```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
normalized_data = StandardScaler().fit_transform(df_market_data[['price_change_percentage_24h', 'price_change_percentage_7d', 'price_change_percentage_14d', 'price_change_percentage_30d', 'price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y']])
```

```python
# Create a DataFrame with the scaled data
normalized_data_df = pd.DataFrame(normalized_data, columns=['price_change_percentage_24h', 'price_change_percentage_7d', 'price_change_percentage_14d', 'price_change_percentage_30d','price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y'])

# Copy the crypto names from the original data
CryptoNames = list(df_market_data.index)

# Set the coinid column as index
normalized_data_df.index = CryptoNames

# Display sample data
normalized_data_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
Clusterlist = [_ for _ in range(1,11)]
```

```python
# Create an empty list to store the inertia values
InertiaList = [] 

# Create a for loop to compute the inertia with each possible value of k

# Inside the loop:

# 1. Create a KMeans model using the loop counter for the n_clusters
for _ in Clusterlist:
    model = KMeans(n_clusters=_, random_state=1)
# 2. Fit the model to the data using `df_market_data_scaled`
    model.fit(normalized_data_df)
# 3. Append the model.inertia_ to the inertia list
    InertiaList.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
InertiaDict = {"ClusterList": Clusterlist, "Inertia": InertiaList}

# Create a DataFrame with the data to plot the Elbow curve
Inertia_df = pd.DataFrame(InertiaDict)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
linechart = Inertia_df.hvplot.line(x="ClusterList", y="Inertia")
linechart
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4 looks to be the best value. Slope begins to notably flatten beyond here.

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
PCAModel=PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
PCAData = PCAModel.fit_transform(normalized_data_df)
PCADatadf = pd.DataFrame(PCAData)

# View the first five rows of the DataFrame. 
PCADatadf.head()
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.

PCAVar = list(PCAModel.explained_variance_ratio_)
PCAVar
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 89.5%. In the above step, this is the sum of the explained variance ratios of the 3 components. .37199 + .34701 + .17604 = .89504 = 89.5%. So it is a good model.

```python
# Create a new DataFrame with the PCA data.
PCADatadf = pd.DataFrame(PCAData, columns=['PCA1', 'PCA2', 'PCA3'])

# Copy the crypto names from the original data
PCADatadf["coin_id"] = df_market_data.index

# Set the coinid column as index
PCADatadf= PCADatadf.set_index("coin_id")

# Display sample data
PCADatadf
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
PCAClusterList = [_ for _ in range(1,11)]
```

```python
# Create an empty list to store the inertia values
PCADataInertiaList = [] 

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:

# 1. Create a KMeans model using the loop counter for the n_clusters
for _ in PCAClusterList:
    model = KMeans(n_clusters=_, random_state=1)
# 2. Fit the model to the data using `df_market_data_pca`
    model.fit(PCADatadf)
# 3. Append the model.inertia_ to the inertia list
    PCADataInertiaList.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
PCAInertiaDict = {'Cluster_List': PCAClusterList, 'Inertia': PCADataInertiaList}

# Create a DataFrame with the data to plot the Elbow curve
PCAInertiaDF = pd.DataFrame(PCAInertiaDict)
PCAInertiaDF.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
PCAPlot = PCAInertiaDF.hvplot.line(x="Cluster_List", y="Inertia")
PCAPlot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4 is the best answer - again this is where the slope notably decreases.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** It's the same, though the inertia is a bit lower.

### Cluster Cryptocurrencies with K-means Using the PCA Data
